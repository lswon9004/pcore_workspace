```python
import pandas as pd, os
os.listdir()
```

```python
CCTV_s = pd.read_csv('CCTV_in_Seoul_2023.csv', encoding='cp949')
CCTV_s.head()
```

```python
CCTV_s
```

```python
CCTV_s = CCTV_s.drop(['Unnamed: 0'],axis=1) # axis = 1 열방향(세로방향)
CCTV_s.head()
```

```python
CCTV_s.fillna(value=0, inplace=True)
CCTV_s.isna()
```

```python
CCTV_s.info()
```

```python
pop_s = pd.read_excel('population_in_Seoul_2023.xlsx',
                     header=2,
                     usecols= 'B,D,G,J')
pop_s.head()
```

```python
pop_s.rename(columns={pop_s.columns[0] : '구별',
                      pop_s.columns[1] : '인구수',
                      pop_s.columns[2] : '한국인',
                      pop_s.columns[3] : '외국인',
                     }, inplace=True)

pop_s.head()
```

```python
CCTV_s.rename(columns={CCTV_s.columns[0] : '구별'}, inplace=True)
CCTV_s.head(2)
```

```python
# CCTV가 제일 많이 설치된 순서대로 정렬(총계)
CCTV_s.sort_values(by='총계', ascending=False)
```

```python
# 최근 증가율 추가
CCTV_s['최근증가율'] = (CCTV_s['2015년']+CCTV_s['2016년']+
                  CCTV_s['2017년']+CCTV_s['2018년']+CCTV_s['2019년']+CCTV_s['2020년']+CCTV_s['2021년']+CCTV_s['2022년']+CCTV_s['2023년']
                  ) / (CCTV_s['2014년 이전']+CCTV_s['2014년'])  * 100

CCTV_s.sort_values(by='최근증가율', ascending=False)
```

```python
CCTV_s.drop([0], inplace=True)
CCTV_s.head()
```

```python
pop_s.drop([0], inplace=True)
pop_s.head()
```

```python
CCTV_s['구별'].unique()
```

```python
pop_s['구별'].unique()
```

```python
CCTV_s.loc[2,'구별'] = '중구'
CCTV_s.loc[2]
```

```python
data_result = pd.merge(CCTV_s, pop_s, on='구별')
data_result
```

```python
data_result.columns
```

```python
del data_result['2014년 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
del data_result['2017년']
del data_result['2018년']
del data_result['2019년']
del data_result['2020년']
del data_result['2021년']
del data_result['2022년']
del data_result['2023년']

data_result.head()
```

```python
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
import numpy as np
np.corrcoef(data_result['인구수'], data_result['총계']) # 상관계수 -1 ~ 1 -1: 강한 반비례, 0:관계없음, 1:강한 비례관계
```

```python
import matplotlib.pyplot as plt

import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
data_result['총계'].plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
data_result['총계'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
data_result['CCTV비율'] = data_result['총계']/ data_result['인구수'] * 100
data_result['CCTV비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['총계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['인구수'],data_result['총계'],1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['총계'], s=50)

plt.plot(fx, f1(fx), ls='dashed', lw = 3, color='g')

plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
data_result['오차'] = np.abs(data_result['총계'] - f1(data_result['인구수']))
df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['총계'], c=data_result['오차'], s = 50)
plt.plot(fx,f1(fx), ls='dashed', lw=3, color="g")

for n in range(25):
    plt.text(df_sort['인구수'][n]* 1.02, df_sort['총계'][n]*0.98,
            df_sort.index[n], fontsize=15)


plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.colorbar()
plt.grid()
plt.show()
```
